```python
#Removing all the special characters like *,+ etc

data['Balls_faced'] = data['Balls_faced'].str.split(pat = '+').str[0]
data['4s'] = data['4s'].str.split(pat = '+').str[0]
data['6s'] = data['6s'].str.split(pat = '+').str[0]
data['Mat'] = data['Mat'].str.split(pat = '*').str[0]
data['Highest_score'] = data['Highest_score'].str.split(pat = '*').str[0]
```

```python
data
```

```python
# Replace '(2024)' with an empty string in the 'Still playing' column
data['Test_Retire'] = data['Test_Retire'].str.replace(r'2024', 'Still_Playing', regex=True)
```

```python
#Changing the datatypes of various columns

data = data.astype({'Mat':'int','Innings':'int','Not_outs':'int','Highest_score':'int','50':'int','4s':'int','6s':'int','Test_Debut':'int'})
data
```
